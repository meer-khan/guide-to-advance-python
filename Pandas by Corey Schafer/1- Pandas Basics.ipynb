# set_option()
pd.set_option() is a useful function to set different options on the dataframe
- to show all rows
- to show all columns 

we set max_rows or max_coulmns to show all, because if we print df normally it turncates rows and columns if they are larger in number, so to see all the data use set_option()

```python
pd.set_option("display.max_columns", 85)
```

```python
# now if we print df, we can see all the 85 columns
df
```
